```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("benchmark_summary.csv")

models = df["model"]

plt.figure()
plt.bar(models, df["avg_relevance"])
plt.title("Average Relevance per Model")
plt.xlabel("Model")
plt.ylabel("Avg Relevance")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("avg_relevance_bar.png", dpi=300)

plt.figure()
plt.bar(models, df["avg_latency"])
plt.title("Average Latency per Model")
plt.xlabel("Model")
plt.ylabel("Latency (seconds)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("avg_latency_bar.png", dpi=300)

plt.figure()
plt.bar(models, df["score"])
plt.title("Final Score per Model")
plt.xlabel("Model")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.savefig("final_score_bar.png", dpi=300)
```

```python
plt.figure()

for i, row in df.iterrows():
    plt.scatter(row["avg_latency"], row["avg_relevance"])
    plt.text(row["avg_latency"], row["avg_relevance"], row["model"])

plt.xlabel("Latency (seconds)")
plt.ylabel("Average Relevance")
plt.title("Trade-off Between Latency and Relevance")

plt.tight_layout()
plt.show()
plt.savefig("tradeoff_scatter.png", dpi=300)
```

```python
df_detail = pd.read_csv("benchmark_detail.csv")

plt.figure()

df_detail.boxplot(column="relevant_docs", by="model")

plt.title("Distribution of Relevant Docs per Model")

plt.suptitle("")

plt.xlabel("Model")

plt.ylabel("Relevant Docs")

plt.tight_layout()

plt.savefig("relevance_boxplot.png", dpi=300)

plt.show()
```

```python
import numpy as np

pivot = df_detail.pivot(index="query", columns="model", values="relevant_docs")

plt.figure()
plt.imshow(pivot, aspect='auto')

plt.colorbar(label="Relevant Docs")

plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45)
plt.yticks(range(len(pivot.index)), pivot.index)

plt.title("Relevance Heatmap (Query vs Model)")

plt.tight_layout()
plt.show()
```
